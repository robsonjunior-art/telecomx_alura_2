```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def avaliar_modelo(nome, modelo, X_train, X_test):
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)

    print(f"\n📊 Modelo: {nome}")
    print("Acurácia:", accuracy_score(y_test, y_pred))
    print("Precisão:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1-score:", f1_score(y_test, y_pred))
    print("Matriz de Confusão:\n", confusion_matrix(y_test, y_pred))
```

```python
lr = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)

avaliar_modelo(
    "Regressão Logística",
    lr,
    X_train_scaled,
    X_test_scaled
)
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

avaliar_modelo(
    "Random Forest",
    rf,
    X_train,
    X_test
)
```

Após a avaliação dos modelos utilizando acurácia, precisão, recall, F1-score e matriz de confusão, observou-se que o Random Forest apresentou o melhor desempenho geral, especialmente em termos de recall, métrica crucial no problema de evasão de clientes. A Regressão Logística apresentou desempenho consistente, porém inferior, enquanto o KNN mostrou maior sensibilidade à escala e ao ruído dos dados. Não foram observados sinais severos de underfitting, porém ajustes de regularização podem ser necessários para evitar overfitting em modelos mais complexos.

1. Introdução

A evasão de clientes (churn) é um dos principais desafios enfrentados por empresas de serviços recorrentes, como telecomunicações. Antecipar quais clientes têm maior probabilidade de evasão permite ações estratégicas de retenção, reduzindo perdas financeiras e aumentando o valor do cliente ao longo do tempo.

Este relatório tem como objetivo:

Identificar os principais fatores que influenciam a evasão

Avaliar o desempenho de diferentes modelos preditivos

Propor estratégias práticas de retenção, com base nos resultados obtidos

2. Descrição e Preparação dos Dados

O conjunto de dados contém informações sobre serviços contratados, formas de pagamento e valores cobrados mensal e cumulativamente, além da variável-alvo Churn, que indica se o cliente evadiu (1) ou permaneceu ativo (0).

2.1 Tratamento dos dados

Variáveis categóricas foram transformadas via One-Hot Encoding

A variável Churn foi convertida para formato binário (0 = não, 1 = sim)

Variáveis numéricas como Charges_Monthly e Charges_Total foram mantidas contínuas
